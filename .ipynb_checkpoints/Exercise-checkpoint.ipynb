```python
from sklearn.model_selection import train_test_split

# Define your target variables
y_ExerciseName = exercise['ExerciseName']
y_ExerciseImage = exercise['ExerciseImage']
y_Equipment = exercise['Equipment']
#y_level = exercise['level']
y_Description = exercise['Description']
y_Duration = exercise['Duration']
y_DaysPerWeek = exercise['DaysPerWeek']
y_TimePerWorkout = exercise['TimePerWorkout']
y_RecommendedSupplements = exercise['RecommendedSupplements']
y_WorkoutPdfwithvideo = exercise['WorkoutPdfwithvideo']

# Define your input features
X = exercise[['Gender', 'Goal','level']]

# Split the data for each target variable
X_train, X_test, y_ExerciseName_train, y_ExerciseName_test, y_ExerciseImage_train, y_ExerciseImage_test, \
y_Equipment_train, y_Equipment_test, y_Description_train, y_Description_test, \
y_Duration_train, y_Duration_test, y_DaysPerWeek_train, y_DaysPerWeek_test, y_TimePerWorkout_train, \
y_TimePerWorkout_test, y_RecommendedSupplements_train, y_RecommendedSupplements_test, y_WorkoutPdfwithvideo_train, \
y_WorkoutPdfwithvideo_test = train_test_split(
    X, y_ExerciseName, y_ExerciseImage, y_Equipment, y_Description, y_Duration, y_DaysPerWeek,
    y_TimePerWorkout, y_RecommendedSupplements, y_WorkoutPdfwithvideo, test_size=0.2, random_state=42)

X_train
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create separate Random Forest Classifier models for each target variable
rf_ExerciseName = RandomForestClassifier()
rf_ExerciseImage = RandomForestClassifier()
rf_Equipment = RandomForestClassifier()
#rf_level = RandomForestClassifier()
rf_Description = RandomForestClassifier()
rf_Duration = RandomForestClassifier()
rf_DaysPerWeek = RandomForestClassifier()
rf_TimePerWorkout = RandomForestClassifier()
rf_RecommendedSupplements = RandomForestClassifier()
rf_WorkoutPdfwithvideo = RandomForestClassifier()

# Fit the models to the respective training data
rf_ExerciseName.fit(X_train, y_ExerciseName_train)
rf_ExerciseImage.fit(X_train, y_ExerciseImage_train)
rf_Equipment.fit(X_train, y_Equipment_train)
#rf_level.fit(X_train, y_level_train)
rf_Description.fit(X_train, y_Description_train)
rf_Duration.fit(X_train, y_Duration_train)
rf_DaysPerWeek.fit(X_train, y_DaysPerWeek_train)
rf_TimePerWorkout.fit(X_train, y_TimePerWorkout_train)
rf_RecommendedSupplements.fit(X_train, y_RecommendedSupplements_train)
rf_WorkoutPdfwithvideo.fit(X_train, y_WorkoutPdfwithvideo_train)

# Make predictions for each target variable
pred_ExerciseName_rf = rf_ExerciseName.predict(X_test)
pred_ExerciseImage_rf = rf_ExerciseImage.predict(X_test)
pred_Equipment_rf = rf_Equipment.predict(X_test)
#pred_level_rf = rf_level.predict(X_test)
pred_Description_rf = rf_Description.predict(X_test)
pred_Duration_rf = rf_Duration.predict(X_test)
pred_DaysPerWeek_rf = rf_DaysPerWeek.predict(X_test)
pred_TimePerWorkout_rf = rf_TimePerWorkout.predict(X_test)
pred_RecommendedSupplements_rf = rf_RecommendedSupplements.predict(X_test)
pred_WorkoutPdfwithvideo_rf = rf_WorkoutPdfwithvideo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Calculate accuracy scores for each target variable
accuracy_ExerciseName_rf = accuracy_score(y_ExerciseName_test, pred_ExerciseName_rf)
accuracy_ExerciseImage_rf = accuracy_score(y_ExerciseImage_test, pred_ExerciseImage_rf)
accuracy_Equipment_rf = accuracy_score(y_Equipment_test, pred_Equipment_rf)
#accuracy_level_rf = accuracy_score(y_level_test, pred_level_rf)
accuracy_Description_rf = accuracy_score(y_Description_test, pred_Description_rf)
accuracy_Duration_rf = accuracy_score(y_Duration_test, pred_Duration_rf)
accuracy_DaysPerWeek_rf = accuracy_score(y_DaysPerWeek_test, pred_DaysPerWeek_rf)
accuracy_TimePerWorkout_rf = accuracy_score(y_TimePerWorkout_test, pred_TimePerWorkout_rf)
accuracy_RecommendedSupplements_rf = accuracy_score(y_RecommendedSupplements_test, pred_RecommendedSupplements_rf)
accuracy_WorkoutPdfwithvideo_rf = accuracy_score(y_WorkoutPdfwithvideo_test, pred_WorkoutPdfwithvideo_rf)
```

```python
import pickle

with open('RandomForestClassifier_ExerciseName.pkl', 'wb') as pickle_out_ExerciseName:
    pickle.dump(rf_ExerciseName, pickle_out_ExerciseName)

with open('RandomForestClassifier_ExerciseImage.pkl', 'wb') as pickle_out_ExerciseImage:
    pickle.dump(rf_ExerciseImage, pickle_out_ExerciseImage)
# Save the RandomForestClassifier model for Equipment
with open('RandomForestClassifier_Equipment.pkl', 'wb') as pickle_out_Equipment:
    pickle.dump(rf_Equipment, pickle_out_Equipment)

# Save the RandomForestClassifier model for level
#with open('RandomForestClassifier_level.pkl', 'wb') as pickle_out_level:
    #pickle.dump(rf_level, pickle_out_level)

# Save the RandomForestClassifier model for Description
with open('RandomForestClassifier_Description.pkl', 'wb') as pickle_out_Description:
    pickle.dump(rf_Description, pickle_out_Description)

# Save the RandomForestClassifier model for Duration
with open('RandomForestClassifier_Duration.pkl', 'wb') as pickle_out_Duration:
    pickle.dump(rf_Duration, pickle_out_Duration)

# Save the RandomForestClassifier model for DaysPerWeek
with open('RandomForestClassifier_DaysPerWeek.pkl', 'wb') as pickle_out_DaysPerWeek:
    pickle.dump(rf_DaysPerWeek, pickle_out_DaysPerWeek)

# Save the RandomForestClassifier model for TimePerWorkout
with open('RandomForestClassifier_TimePerWorkout.pkl', 'wb') as pickle_out_TimePerWorkout:
    pickle.dump(rf_TimePerWorkout, pickle_out_TimePerWorkout)

# Save the RandomForestClassifier model for RecommendedSupplements
with open('RandomForestClassifier_RecommendedSupplements.pkl', 'wb') as pickle_out_RecommendedSupplements:
    pickle.dump(rf_RecommendedSupplements, pickle_out_RecommendedSupplements)

# Save the RandomForestClassifier model for WorkoutPdfwithvideo
with open('RandomForestClassifier_WorkoutPdfwithvideo.pkl', 'wb') as pickle_out_WorkoutPdfwithvideo:
    pickle.dump(rf_WorkoutPdfwithvideo, pickle_out_WorkoutPdfwithvideo)



# Close the files
pickle_out_ExerciseName.close()
pickle_out_ExerciseImage.close()
pickle_out_Equipment.close()
#pickle_out_level.close()
pickle_out_Description.close()
pickle_out_Duration.close()
pickle_out_DaysPerWeek.close()
pickle_out_TimePerWorkout.close()
pickle_out_RecommendedSupplements.close()
pickle_out_WorkoutPdfwithvideo.close()


with open('exercise.pkl', 'wb') as pickle_out_exercise:
    pickle.dump(exercise, pickle_out_exercise)

# Close the files
pickle_out_exercise.close()
```
